# Linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

## Load data
df = pd.read_csv('melb_data_sold_train.csv')
```

## Remove rows with missing attributes

```python
df1 = df.dropna().reset_index()
```

## Add [Haversine distance](https://en.wikipedia.org/wiki/Haversine_formula) attribute

```python
# pt1 and pt2 are 2x1 [lat,long] np arrays that contain locations of the 2 earth coordinates in deg
def haversine_distance(pt1, pt2):
    EARTH_RADIUS = 6371.0088
    # convert to radians
    pt1 = np.radians(pt1)
    pt2 = np.radians(pt2)
    # calulations
    radicand = ((np.sin((pt1[0] - pt2[0]) / 2)) ** 2) + np.cos(pt1[0]) * np.cos(pt2[0]) * (np.sin((pt2[1] - pt1[1]) / 2)) ** 2
    return 2 * EARTH_RADIUS * np.arcsin(radicand ** .5)
```

```python
# Check function
toorak_pt = np.array([-37.841820, 145.015986])
haversine_distance(toorak_pt,[-37.68178,144.73779])
```

```python
df2 = df1.copy()
df2['distance_to_toorak'] = haversine_distance(toorak_pt, np.array([df2['Lattitude'], df2['Longtitude']]))
df2
```

## Creating one hot encoding for 'Type'

```python
df3 = df2.copy()
type_dummy = pd.get_dummies(df3['Type'])
del df3['Type']
df3 = df3.join(type_dummy)
df3
```

## Calculate pairwise correlations

```python
df3.corr()
```

## Create a linear regression model for price

```python
# X = [1s, h, t, u, distance]
ones = pd.DataFrame({'intercept': np.ones(len(df3))})
X = ones.join([df3['h'], df3['t'], df3['u'], df3['distance_to_toorak']])
X
```

```python
y = df3['Price']
y
```

```python
# get beta_hat
[beta_hat, residuals, rank, s] = np.linalg.lstsq(X, y, rcond = -1)
beta_hat
```

```python
# Test model
np.matmul(X.iloc[23], beta_hat) 
```

```python
# Actual price
y[23]
```

## Appling linear regression model and visualizing error

```python
test_df = pd.read_csv('melb_data_sold_test.csv')
test_df.fillna('na')
```

```python
test_df['distance_to_toorak'] = test_df.apply(lambda o : haversine_distance(toorak_pt, (o['Lattitude'], o['Longtitude'])), 
                                              axis = 1)
test_type_dummy = pd.get_dummies(test_df['Type'])
del test_df['Type']
test_df = test_df.join(test_type_dummy)
```

```python
test_ones = pd.DataFrame({'intercept': np.ones(len(test_df))})
Xdot = test_ones.join([test_df['h'], test_df['t'], test_df['u'], test_df['distance_to_toorak']])
Xdot_arr = Xdot.to_numpy()
```

```python
y_hat = np.matmul(Xdot_arr, beta_hat)
```

```python
# Calculate absolute error
error = abs(test_df['Price'] - y_hat)
```

```python
# Creating a histogram of the absolute error
bins = [x for x in range(0, 6_000_000, 100_000)]
plt.hist(error, alpha = 1, label = 'error', color = '#89CFF0', bins = bins)
plt.hist(test_df['Price'], alpha = .5, color = 'pink', label = 'actual price', bins = bins)
plt.title('Actual property price and absolute error')
plt.xlabel('Price (millions AUD)')
plt.ylabel('Number of properties')
plt.legend(loc = 'upper right')
plt.show()
```
