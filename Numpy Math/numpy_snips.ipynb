## Discrete PDF
[numpy random normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

```python
nums = np.random.normal(5, 10, 1000)
plt.hist(nums, 41)
plt.xlabel('Value')
plt.ylabel('Number of generated values')
plt.title('PDF of quasi-random values')
```

## Discrete CDF

```python
# Make plot
plt.hist(nums, 40, density=True, histtype='step', cumulative=True)
x = np.percentile(nums, 80)
point, = plt.plot(x, .8, 'g*')
plt.xlim(min(nums), max(nums))
plt.xlabel('Value')
plt.ylabel('Cumulative probability')
plt.title('CDF of quasi-random values')
plt.legend(['80th percentile'])
```

## Rotation with Matrices


```python
# Make matrices.
# [[x1, x2], 
#  [y1, y2]]

alpha = np.array([[0, 1],
                  [0, 1]])

beta = np.array([[1, 0],
                 [1, 1]])

angle = np.pi / 4
rotation_m = np.array([[np.cos(angle), -np.sin(angle)],
                       [np.sin(angle), np.cos(angle)]])

rotated_alpha = np.matmul(rotation_m, alpha)
rotated_beta = np.matmul(rotation_m, beta)
```

```python
# Draw graph
ax = plt.axes()

# Alpha
ax.arrow(alpha[0][0], alpha[1][0], alpha[0][1], alpha[1][1], ec='teal')
ax.arrow(rotated_alpha[0][0], rotated_alpha[1][0], rotated_alpha[0][1], rotated_alpha[1][1], ec='teal')

# Beta
ax.arrow(beta[0][0], beta[1][0], beta[0][1], beta[1][1], ec='purple')
ax.arrow(rotated_beta[0][0], rotated_beta[1][0], rotated_beta[0][1], rotated_beta[1][1], ec='purple')

plt.grid()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title('Rotating Vector')
plt.xlabel('x')
plt.ylabel('y')
```
