# Clustering with DBSCAN

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

# DBSCAN

## Test K-means
Using:
[K-means](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```python
def random_color():
    color = ["#"+''.join([np.random.choice([c for c in '0123456789ABCDEF']) for j in range(6)])]
    return color[0]
```

```python
df_dbscan = pd.read_csv('data_dbscan.csv')

# Get pairs 
points = [[p['x'], p['y']] for i, p in df_dbscan.iterrows()]

# Call k_means with k = 2
kmeans = KMeans(n_clusters = 2)
labels = kmeans.fit_predict(points)

# Plot with colors
colors = [random_color() for i in np.unique(labels)]
color_ls = [colors[i] for i in labels]
plt.scatter(df_dbscan['x'], df_dbscan['y'], color = color_ls)
```

## Range Query Function


```python
def range_query( df_data, q_index, eps ):
    neighbors = pd.DataFrame()
    for i, p in df_data.iterrows():
        if np.linalg.norm(p - df_data.iloc[q_index]) <= eps:
            neighbors = neighbors.append(p)
    return neighbors
```

```python
# Test
df_n = range_query( df_dbscan, 0, .1 )
plt.plot( df_dbscan['x'], df_dbscan['y'], '.', alpha = 0.5 )
plt.plot( df_n['x'], df_n['y'], 'r*' )
plt.xlabel('x')
plt.ylabel('y')
plt.title('Testing Range Query')
plt.show()
```

## DBSCAN Function

```python
def dbscan( df_data, eps, min_pts ):
    NOISE_LABEL = 0
    clusterCount = 0
    labels = pd.Series(['*' for i in range(len(df_data))])
    # For each point 
    for i, p in df_data.iterrows():
        
        # If already labeled, move to next point
        if labels[i] != '*':
            continue
                
        # Get neighbors
        neighbors = range_query(df_data, i, eps)
        
        # Check if meets density reqs, if not, stop and label as noise 
        if len(neighbors) < min_pts:
            labels[i] = NOISE_LABEL
            continue
            
        clusterCount += 1
        
        # Set label of point 
        labels[i] = clusterCount
        
        # get set(all neighbors - current point) to expand on 
        # S = [[org_index, series(x, y)], [org_index, series(x, y)], ...]
        S = []
        for j, n in neighbors.iterrows():

            if not n.equals(p):
                S.append([j, n])
                
        # for each neighbor in set S
        for point in S:
            
            # if not (*) label, then continue
            if labels[point[0]] != '*':
                continue
            
            # if noise (so border) or unlabeled, change to border point
            # add cluster label
            labels[point[0]] = clusterCount
            
            # get neighbors 
            inner_neighbors = range_query(df_data, point[0], eps)
            
            # if num neighbors >= minPoints, add to set S
            if len(inner_neighbors) >= min_pts:
                
                # for each inner_neighbor
                for idx, inner_n in inner_neighbors.iterrows():
                    
                    # check if already in S using index
                    if idx not in [i[0] for i in S]:
                        
                        # if not, append to S
                        S.append([idx, inner_n])

    return labels
```

```python
# Test it out
s_cluster_assignment = dbscan(df_dbscan, 0.23, 6)
s_cluster_assignment.value_counts()
```

## Plot Results 

```python
# Plot each cluster
colors = [random_color() for i in np.unique(s_cluster_assignment)]
color_ls = [colors[i] for i in s_cluster_assignment]
plt.scatter(df_dbscan['x'], df_dbscan['y'], color = color_ls)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Custom DBScan Results')
plt.show()
```

## Compare to scikit-learn implementation

```python
# check your implementation vs scikit-learn
from sklearn.cluster import DBSCAN
labels = DBSCAN(eps = .23, min_samples = 6).fit(df_dbscan.to_numpy()).labels_

colors = [random_color() for i in np.unique(labels)]
color_ls = [colors[i] for i in labels]
plt.scatter(df_dbscan['x'], df_dbscan['y'], color = color_ls)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sklearn DBScan Results')
plt.show()
```
